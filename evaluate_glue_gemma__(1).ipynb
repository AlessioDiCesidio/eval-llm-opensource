```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from datasets import load_metric, load_dataset
from tqdm import tqdm

# Define model name (Mistral 7B) and task name
model_name = "google/gemma-7b"
task_name = "mnli"

# Load tokenizer and model
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Load Glue metric
metric = load_metric("glue", "mnli")

###on test matched
def evaluate_mnli(glue_dataset):
  """Evaluates MNLI predictions with Glue metrics and progress bar"""
  predictions, labels = [], []
  total_examples = len(glue_dataset)

  with tqdm(total=total_examples, desc="Evaluating MNLI Dev Set") as progress_bar:
    for inputs in glue_dataset:
      premise = inputs["premise"]
      hypothesis = inputs["hypothesis"]

      # Tokenize premise and hypothesis
      encoding = tokenizer(premise, hypothesis, return_tensors="pt")

      # Forward pass through model
      outputs = model(**encoding)
      predictions.append(outputs.logits.argmax(-1).item())
      labels.append(inputs["label"])
      progress_bar.update(1)  # Update progress bar for each example

  # Evaluate predictions
  results = metric.compute(predictions=predictions, references=labels)
  return results


# Load MNLI dataset from Glue
glue_dataset = load_dataset("kmyoo/glue-mnli-tiny")

# Evaluate model on MNLI dev set
dev_results = evaluate_mnli(glue_dataset["test_matched"])
print(f"MNLI Dev Results: {dev_results}")
```

```python


glue_dataset_qnli = load_dataset("glue", "qnli")
metric_qnli=load_metric("glue","qnli")
sampled_dataset = glue_dataset_qnli["validation"].select(range(100))

def evaluate_qnli(glue_dataset_qnli):
  """Evaluates MNLI predictions with Glue metrics and progress bar"""
  predictions, labels = [], []
  total_examples = len(glue_dataset_qnli)

  with tqdm(total=total_examples, desc="Evaluating QNLI Dev Set") as progress_bar:
    for inputs in glue_dataset_qnli:
      entailment = inputs["question"]
      not_entailment = inputs["sentence"]

      # Tokenize premise and hypothesis
      encoding = tokenizer(entailment, not_entailment, return_tensors="pt")

      # Forward pass through model
      outputs = model(**encoding)
      predictions.append(outputs.logits.argmax(-1).item())
      labels.append(inputs["label"])
      progress_bar.update(1)  # Update progress bar for each example

  # Evaluate predictions
  results = metric_qnli.compute(predictions=predictions, references=labels)
  return results


# Evaluate model on the sampled validation set
dev_results_qnli = evaluate_qnli(sampled_dataset)
print(f"QNLI Dev Results on 100 Samples: {dev_results_qnli}")
```
